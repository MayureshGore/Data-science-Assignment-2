```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = {
    'Name of company': ['Allied Signal', 'Bankers Trust', 'General Mills', 'ITT Industries', 'J.P.Morgan & Co.',
                        'Lehman Brothers', 'Marriott', 'MCI', 'Merrill Lynch', 'Microsoft', 'Morgan Stanley',
                        'Sun Microsystems', 'Travelers', 'US Airways', 'Warner-Lambert'],
    'Measure X': [24.23, 25.53, 25.41, 24.14, 29.62, 28.25, 25.81, 24.39, 40.26, 32.95, 91.36, 25.99, 39.42, 26.71, 35.00]
}
```

```python
df = pd.DataFrame(data)
```

```python
plt.figure(figsize=(12, 6))
plt.bar(df['Name of company'], df['Measure X'], color='skyblue')
plt.xticks(rotation=90)
plt.title('Measure X by Company')
plt.xlabel('Company')
plt.ylabel('Measure X (%)')
plt.show()
```

```python
mean = np.mean(df['Measure X'])
std_dev = np.std(df['Measure X'])
variance = np.var(df['Measure X'])

print("Mean (μ):", mean)
print("Standard Deviation (σ):", std_dev)
print("Variance (σ^2):", variance)
```

```python
outliers = df[(df['Measure X'] > mean + 3 * std_dev) | (df['Measure X'] < mean - 3 * std_dev)]
print("Outliers:")
print(outliers)
```
